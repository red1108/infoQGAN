```python
# Standard Libraries
import math
import pickle
import random
import numpy as np

# 데이터 분포를 만들었던 코드를 저장하기 위한 모듈
# 어떤 함수의 내부 코드를 불러올 수 있게 하는 모듈이다.
import inspect
```

```python
import importlib
import matplotlib.pyplot as plt
from modules import PointGenerator  # 초기 import
importlib.reload(PointGenerator)  # 모듈 갱신

# 클래스 인스턴스 생성
data_num = 1000
data_type = "biased_box2"
idx = 1
generator = PointGenerator.PointGenerator(data_num)
```

```python
# 원하는 분포 생성 (예: 'curve')

generate_function = generator.generate_biased_box2 # 이걸 바꿔주면 된다.
data = generate_function()

# 그래프 그리기
plt.figure(figsize=(5, 5))
plt.scatter(data[:, 0], data[:, 1], s=2.0)
plt.title('Target distribution')
plt.xlim((0, 1))
plt.ylim((0, 1))

np.savetxt(f'data/2D/{data_type}_{data_num}_{idx}.txt', data) # 2차원 점 분포를 txt 파일에 저장
plt.savefig(f'data/2D/{data_type}_{data_num}_{idx}.png') # 2차원 점 분포를 png 파일로 저장
plt.show()

function_code = inspect.getsource(generate_function)
output_code_path = f'data/2D/{data_type}_{data_num}_{idx}_code.txt'
with open(output_code_path, 'w') as file:
    file.write(function_code)
```
