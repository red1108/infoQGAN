D_ks, p_value, mi, angle, code1_norm, code2_norm 을 평가하자.

가장 p_value 가 높은 epoch 기준으로 평가하면 된다.

1. csv 파일을 불러들여서 p_value 가 가장 높은 epoch를 찾고
2. 해당 epoch의 numpy/gen_outputs_epoch_{epoch}.txt 에서 점들을 불러온다
3. 해당 epoch의 numpy/gen_codes_epoch_{epoch}.txt 에서 code를 불러온다
4. 그리고 위 값들을 평가하고, pandas 로 결과 만들어서 이쁘게 보여준다.

```python
import numpy as np
import pandas as pd

# 대상 디렉터리와 모델 이름 리스트
dirs = [
    ["./정리/Scientific Reports/2D/GAN_sd5_hd5_biased_diamond_Mar15_22_47_57", "GAN"],
    ["./정리/Scientific Reports/2D/GAN_sd5_hd4_biased_diamond_Mar15_23_39_46", "GAN"],
    ["./정리/Scientific Reports/2D/InfoGAN_sd5_hd5_biased_diamond_Mar15_22_47_59", "InfoGAN"],
    ["./정리/Scientific Reports/2D/InfoGAN_sd5_hd4_biased_diamond_Mar15_23_39_41", "InfoGAN"],
    ["./정리/Scientific Reports/2D/True_np5_nl10_biased_diamond_True_Mar08_18_52_11", "InfoQGAN"],
    ["./정리/Scientific Reports/2D/True_np5_nl10_biased_diamond_True_Mar08_18_52_16", "InfoQGAN"],
    ["./정리/Scientific Reports/2D/False_np5_nl10_biased_diamond_False_Mar08_18_52_22", "QGAN"],
    ["./정리/Scientific Reports/2D/False_np5_nl10_biased_diamond_False_Mar08_18_52_18", "QGAN"],

]

# 결과를 저장할 DataFrame 생성
result = pd.DataFrame(columns=["D_ks", "p_value", "mi", "angle", "code1_norm", "code2_norm", "model"])

for directory, model in dirs:
    # 각 디렉터리의 scalars.csv 파일 읽기
    df = pd.read_csv(f"{directory}/scalars.csv")
    
    # p_value가 가장 높은 행 찾기
    max_row = df.loc[df["p_value"].idxmax()]
    
    # code1_norm 계산: (Corr/code0-y, Corr/code0-x) 벡터의 노름
    code1_norm = np.sqrt(max_row["Corr/code0-y"]**2 + max_row["Corr/code0-x"]**2)
    
    # code2_norm 계산: (Corr/code1-y, Corr/code1-x) 벡터의 노름
    code2_norm = np.sqrt(max_row["Corr/code1-y"]**2 + max_row["Corr/code1-x"]**2)
    
    # 필요한 값들을 하나의 딕셔너리로 정리
    new_data = {
        "D_ks": max_row["D_ks"],
        "p_value": max_row["p_value"],
        "mi": max_row["MI"],
        "angle": max_row["angle"],
        "code1_norm": code1_norm,
        "code2_norm": code2_norm,
        "model": model
    }
    
    # 새로운 DataFrame 생성 후 pd.concat 사용하여 결합
    new_df = pd.DataFrame([new_data])
    result = pd.concat([result, new_df], ignore_index=True)

# 최종 결과 출력
print(result)
```
